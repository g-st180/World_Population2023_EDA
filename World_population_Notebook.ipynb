# Waffle Chart

```python
filtr=df['Worldpercentage']<1
df.loc[filtr,'Worldpercentage'].sum()
```

```python
filtr=df['Worldpercentage']>1
temp_df=df.loc[filtr]
temp_df
```

```python
temp_df=temp_df.append({'Country': 'Others','Worldpercentage':31.319}, ignore_index=True)
```

```python
temp_df
```

```python
fig=plt.figure(FigureClass=Waffle,
               rows=9,
               columns=50,
               figsize=(18,8),
              values=temp_df['Worldpercentage'],
              labels=list(temp_df['Country']),
              legend={'loc':(0,-0.7),'ncol':8,'fontsize':15,'framealpha':0})
```
